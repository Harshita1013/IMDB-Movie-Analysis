# Outlier Calculation

```python
med = df['score'].median()
print(med)
```

```python
q1, q3 = df['score'].quantile((0.25, 0.75))
print(q1)
print(q3)
```

```python
IQR = q3-q1
print(IQR)
```

```python
lower_fence = q1 - 1.5*IQR
upper_fence = q3 + 1.5*IQR
print(lower_fence)
print(upper_fence)
```

```python
print(f'Median: {med}\nQuartile 1(25%): {q1}\nQuartile 3(75%): {q3}\nIQR: {IQR}\nLower Bound: {lower_fence}\nUpper Bound: {upper_fence}')
```

```python
outliers = df[(df['score'] < lower_fence) | (df['score'] > upper_fence)]
outliers
```

```python
#Histogram Describes the negative skewness only in 2022 the demand of the movie will high
import seaborn as sns
import matplotlib.pyplot as plt


df = df.sort_values(by = 'Year')
plt.figure(figsize = (12,4))
sns.histplot(df['Year'], bins = 20, kde = True)
plt.title('Distribution Of Year')
plt.ylabel('Frequency')
plt.xticks(rotation = 90, fontsize = 6)
plt.show()
```
